```python
import cv2
import numpy as np
from skimage import io

def adjust_gamma(image, gamma=1.0):
# image, gamma=1.0
   invGamma = 1.0 / gamma
   table = np.array([((i / 255.0) ** invGamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

   return cv2.LUT(image, table)

x = 'C:/Users/SARA/Documents/IMAGENES/1gamma.jpg'  #location of the image
original = cv2.imread(x, 1)
gamma = 5     # change the value here to get different result
adjusted = adjust_gamma(original, gamma=gamma)
cv2.putText(adjusted, "g={}".format(gamma), (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
io.imshow(adjusted)
```

```python
import cv2
import numpy as np
from skimage import io

def adjust_gamma(image, gamma=1.0):
# image, gamma=1.0
   invGamma = 1.0 / gamma
   table = np.array([((i / 255.0) ** invGamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

   return cv2.LUT(image, table)

x = 'C:/Users/SARA/Documents/IMAGENES/2gamma.jpg'  #location of the image
original = cv2.imread(x, 1)
gamma = 4       # change the value here to get different result
adjusted = adjust_gamma(original, gamma=gamma)
cv2.putText(adjusted, "g={}".format(gamma), (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
io.imshow(adjusted)
```
