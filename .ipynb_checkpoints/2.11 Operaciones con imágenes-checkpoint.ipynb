```python
# Restar las imágenes de manera que se detecte las regiones que se han alterado en cada
# imagen, esto lo puede realizar usando la operación simple de resta entre matrices.

import cv2
from skimage.color import rgb2gray
from skimage import io

 
#Cargamos las dos imagenes para hacer las diferencias
diff1 = cv2.imread('abue1.JPG')
diff2 = cv2.imread('abue2.JPG')
diff_total = cv2.absdiff(diff1, diff2)
io.imshow(diff_total)
print(diff1.shape)


```

```python
#Implementar una función que recorra cada pixel en las imágenes y realice la operación
#resta. Comparar el resultado con el obtenido en el caso anterior.

from skimage import io 
im1 = cv2.imread('abue1.JPG')
im2 = cv2.imread('abue2.JPG')
im3=im2
print(im1.shape)
i=0
while i < im1.shape[0]:
    j = 0
    while j < im1.shape[1]:
        im3[i][j]  = im1[i][j] - im2[i][j]
        j+=1 
    i+=1
io.imshow(im3)
io.imsave('cerebro_ng.JPG', im1)
```
