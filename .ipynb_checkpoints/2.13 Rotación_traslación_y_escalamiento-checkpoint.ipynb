Implemente una función que realice las tres funciones anteriores para ser usadas de manera simple en la imagen recomendada y

• rotar el triángulo 95 grados

https://python-para-impacientes.blogspot.com/2014/12/fundamentos-para-procesar-imagenes-con.html




```python
from PIL import Image
imagen = Image.open('Imagenes_Pruebas/FIGURAS_salida.jpg')
# Mostrar tanaño de la imagen:
imagen.size  # 399x200

print(imagen.size)


# Definir tupla con las coordenadas de la región
caja = (280, 0, 410, 130)

# Obtener región 
region = imagen.crop(caja)

# Girar región 90º
region = region.transpose(Image.ROTATE_90)

# Pegar región girada en la imagen original
imagen.paste(region, caja)

# Mostrar imagen final y guardar
imagen.show()
imagen.save("Imagenes_Pruebas/regiongirada90.jpg")
```

• Trasladar el rectángulo a la derecha de su posición una cantidad que se aprecie.

```python

import cv2
import numpy as np

import scipy.misc

img = cv2.imread('Imagenes_Pruebas/FIGURAS_salida.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
scipy.misc.imsave('Imagenes_Pruebas/trasladada.jpg', dst)
```
