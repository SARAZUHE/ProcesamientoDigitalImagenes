Restar las imágenes de manera que se detecte las regiones que se han alterado en cada imagen, esto lo puede realizar usando la operación simple de resta entre matrices.
Implementar una función que recorra cada pixel en las imágenes y realice la operación resta. Comparar el resultado con el obtenido en el caso anterior.
Como pudo observar las imágenes deben ser alineadas para que el resultado de la resta pueda permitir detectar las regiones que hacen la diferencia en cada par de imágenes. Implemente una función de alineación en cada par de imágenes; una vez alineadas repita el proceso de resta y debe obtener una imagen binaria donde las regiones alteradas deben aparecer como zonas blancas.

```python

 
from PIL import Image
 
#carga de las imagenes
 
imagen1 = Image.open('Imagenes_Pruebas/hw1_painting_1_reference.jpg').convert('L')
 
imagen2 = Image.open('Imagenes_Pruebas/hw1_painting_1_tampered.jpg').convert('L')
 
#obtencion de los datos de ambas imagenes
 
datos_imagen1 = imagen1.getdata()
 
datos_imagen2 = imagen2.getdata()
 
#calculo de diferencias
 
datos_diferencias = [abs(datos_imagen1[x]-datos_imagen2[x]) for x in range(len(datos_imagen1))]
 
#creacion de la nueva imagen con las diferencias mapeadas
 
imagen_diferencias = Image.new('L', imagen1.size)
 
imagen_diferencias.putdata(datos_diferencias)
 
imagen_diferencias.save('Imagenes_Pruebas/salidahw1_painting_1_tampered.jpg')
 
imagen1.close()
imagen2.close()
 
imagen_diferencias.close()
```
