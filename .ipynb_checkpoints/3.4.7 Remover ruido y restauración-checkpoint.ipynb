```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image = cv2.imread('original.JPG')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gaussian_3 = cv2.GaussianBlur(gray_image, (5,5), 10.0)
unsharp_image = cv2.addWeighted(gray_image, 1.5, gaussian_3, -0.5, 0, image)


plt.subplot(121),plt.imshow(image,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(gaussian_3,cmap = 'gray')
plt.title('Gaussian Image'), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(121),plt.imshow(unsharp_image,cmap = 'gray')
plt.title('Unsharp Image'), plt.xticks([]), plt.yticks([])




```
