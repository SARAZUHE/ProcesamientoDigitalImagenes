```python
from PIL import Image, ImageFilter

desviacion = 10
imagen_original = Image.open('Imagenes_Pruebas/ima_medica.tif')

plt.subplot(121),plt.imshow(imagen_original),plt.title('Original')
plt.xticks([]), plt.yticks([])
imagen_filtrada = imagen_original.filter(ImageFilter.GaussianBlur(desviacion))
 
plt.subplot(122),plt.imshow(imagen_filtrada),plt.title('Filtro Gausiano')
plt.xticks([]), plt.yticks([])
imagen_filtrada.save('Imagenes_Pruebas/ima_medica_2.tif')
 
imagen_original.close()
 
imagen_filtrada.close()
```
